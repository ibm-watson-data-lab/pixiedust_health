# Initialize App

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import pixiedust
from pixiedust_health import cohort, pixieHealthApp
```

# Load Data

```python
path_to_synthetic_data = './synthetic_data/'

demographics = pd.read_csv(path_to_synthetic_data + 'create_v_demographic.csv', delimiter="\t")
diagnosis = pd.read_csv(path_to_synthetic_data + 'create_v_diagnosis.csv', delimiter="\t")
observations = pd.read_csv(path_to_synthetic_data + 'create_v_observation.csv', delimiter="\t")
```

# Create Cohorts class

```python
chrt = cohort.Cohorts(demographics, diagnosis, observations)
display(chrt.demographicsFeatures)
```

## Descriptive Analytics

```python
disease = 1
tmp = chrt.getDemographics(disease)
display(tmp['pos'])
```

## Classification Reports

```python
allFeatures = chrt.getFeatureVectors(disease)
print("Features: ", chrt.getFeatureNames(allFeatures))
```

```python
clf_report, roc_df, feat_df = chrt.getClassification(allFeatures)
```

```python
display(clf_report)
```

```python
pixieHealthApp.PixieHealthApp().run(chrt, runInDialog='true')
```
